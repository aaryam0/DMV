```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sales_data_sample.csv',encoding='latin-1')
```

```python
df.head()
```

```python
df.describe()
```

```python
newdf = df.fillna(0)
```

```python
newdf['LOCATION'] = newdf['CITY'] + ',' + newdf['COUNTRY']
```

```python
total_sales = newdf['SALES'].sum()
total_sales
```

```python
avg_order_value = newdf['SALES'].mean()
avg_order_value
```

```python
product_category = newdf['PRODUCTLINE'].value_counts()
product_category
```

```python
newdf.head()
```

```python
product_category.plot(kind='bar')
plt.xlabel("product")
plt.ylabel("sales")
plt.title("product category distribution")
plt.show()
```

```python
product_category.plot(kind='pie',autopct='%1.0f%%')
plt.legend(loc = 0, bbox_to_anchor=(1.2, 1.1))
plt.title("product category distribution")
plt.show()
```

```python
product_category.plot(kind='box')
plt.xlabel("product")
plt.ylabel("sales")
plt.title("product category distribution")
plt.show()
```
